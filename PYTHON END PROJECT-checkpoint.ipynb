```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("abc_company.csv")
print(df)
#Data Exploration
print("Data Exploration:")
print(df.info())
print("\n\nSample Data:")
print(df.head())
print("\n\nNull values in columns:")
print(df.isnull().sum())


```

```python
df['Height'] = np.random.randint(150, 181, size=len(df))
print(df['Height'])
```

```python
#TASK 1 .Distribution of Employees Across Each Team:
team_distribution = df['Team'].value_counts(normalize=True) * 100
print(team_distribution)
#We can use a bar plot to show the percentage split of employees across different teams.
plt.figure(figsize=(50, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values)
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Percentage of Employees')
plt.grid(axis="y",linestyle="--")
plt.show()
```

```python
#TASK 2.Segregation of Employees Based on Their Positions
position_distribution = df['Position'].value_counts()
print(position_distribution)
#We can use a pie chart or a bar plot to show the percentage split of employees based on positions.
plt.figure(figsize=(8, 8))
plt.pie(position_distribution, labels=position_distribution.index, autopct='%1.1f%%')
plt.title('Segregation of Employees by Position')
plt.axis('equal')
plt.show()
```

```python
# TASK 3.Predominant Age Group Among Employees:
age_distribution = pd.cut(df['Age'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
predominant_age_group = age_distribution.value_counts().idxmax()
print(predominant_age_group)
#We can use a count plot to visualize the distribution of employees across different age groups.
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x=age_distribution)
plt.title('Distribution of Employees by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# TASK 4.Team and Position with the Highest Salary Expenditure:
highest_salary_team = df.groupby('Team')['Salary'].sum().idxmax()
highest_salary_position = df.groupby('Position')['Salary'].sum().idxmax()
print(f"Team with highest salary expenditure: {highest_salary_team}")
print(f"Position with highest salary expenditure: {highest_salary_position}")
# Calculate total salary expenditure by TEAM
total_salary_team = df.groupby('Team')['Salary'].sum()
# Plotting total salary expenditure across teams
plt.figure(figsize=(12, 6))
plt.bar(total_salary_team.index, total_salary_team.values, color='skyblue')
plt.title('Total Salary Expenditure Across Teams')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
# Calculate total salary expenditure by POSITI
total_salary_position = df.groupby('Position')['Salary'].sum()
# Plotting total salary expenditure by position
plt.figure(figsize=(10, 6))
plt.bar(total_salary_position.index, total_salary_position.values, color='salmon')
plt.title('Total Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#TASK 5.Correlation Between Age and Salary:
#We can use a scatter plot to visualize the relationship between age and salary.
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Salary')
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
